# 01-SQL-Consultas-Básicas

```python
# Librerias
import pandas as pd
from sqlalchemy import create_engine
```

```python
# Conerctar a mi SQLserver
driver = "ODBC Driver 17 for SQL Server"
server = r"DESKTOP-GB9FFBV"
database = "BD_SQL_"
```

```python
# Crear motor de conexión
connection_string = f"mssql+pyodbc://@{server}/{database}?driver={driver}"
engine = create_engine(connection_string)
```

```python
# -------------------------------
# 1. Mostrar todos los registros de la tabla clientes.
# -------------------------------
df1 = pd.read_sql("SELECT * FROM clientes;", engine)
print(df1)
```

```python
# -------------------------------
# 2. Listar los nombres de todos los productos disponibles.
# -------------------------------
df2 = pd.read_sql("SELECT id_producto, nombre_producto FROM productos;", engine)
print(df2)
```

```python
# -------------------------------
# 3. Obtener los nombres y salarios de todos los empleados.
# -------------------------------
df3 = pd.read_sql("SELECT id_vendedor, nombre, salario FROM empleados;", engine)
print(df3)
```

```python
# -------------------------------
# 4. Mostrar las ventas realizadas en el año 2024.
# -------------------------------
df4 = pd.read_sql("SELECT * FROM ventas WHERE DATEPART(YEAR, fecha_venta) = 2024;", engine)
print(df4)
```

```python
# -------------------------------
# 5. Listar los clientes registrados en el último mes.
# -------------------------------
df5 = pd.read_sql("""
SELECT *
FROM clientes
WHERE fecha_registro >= DATEADD(month,-1,DATETRUNC(month, SYSDATETIME()))
  AND fecha_registro < DATETRUNC(month, SYSDATETIME());
""", engine)
print(df5)
```

```python
# -------------------------------
# 6. Calcular el total de ingresos (cantidad * precio_unitario) de la tabla ventas.
# -------------------------------
df6 = pd.read_sql("SELECT SUM(precio_unitario * cantidad) AS total_ingresos FROM ventas;", engine)
print(df6)
```

```python
# -------------------------------
# 7. Contar cuántas ventas se registraron por canal de venta.
# -------------------------------
df7 = pd.read_sql("SELECT canal, COUNT(id_venta) AS total_ventas FROM ventas GROUP BY canal;", engine)
print(df7)
```

```python
# -------------------------------
# 8. Mostrar los productos vendidos en la región 'Lima'.
# -------------------------------
df8 = pd.read_sql("""
SELECT v.id_producto, p.nombre_producto
FROM ventas v
LEFT JOIN productos p ON v.id_producto = p.id_producto
WHERE v.region = 'Lima';
""", engine)
print(df8)
```

```python
# -------------------------------
# 9. Obtener el número total de clientes por país.
# -------------------------------
df9 = pd.read_sql("SELECT pais, COUNT(id_cliente) AS total_clientes FROM clientes GROUP BY pais;", engine)
print(df9)
```

```python
# -------------------------------
# 10. Mostrar los 10 clientes más recientes (ordenados por fecha_registro).
# -------------------------------
df10 = pd.read_sql("""
SELECT TOP (10) id_cliente, nombre, fecha_registro
FROM clientes
ORDER BY fecha_registro DESC;
""", engine)
print(df10)
```

```python
# -------------------------------
# 11. Listar las ventas donde la cantidad vendida sea mayor a 4 unidades.
# -------------------------------
df11 = pd.read_sql("SELECT * FROM ventas WHERE cantidad > 4 ORDER BY cantidad DESC;", engine)
print(df11)
```

```python
# -------------------------------
# 12. Calcular el precio de venta promedio por categoría de producto.
# -------------------------------
df12 = pd.read_sql("""
SELECT p.categoria, AVG(v.precio_unitario) AS precio_unit_promedio
FROM productos p LEFT JOIN ventas v ON p.id_producto = v.id_producto
GROUP BY p.categoria;
""", engine)
print(df12)
```

```python
# -------------------------------
# 13. Mostrar los empleados cuyo salario es superior a 3000.
# -------------------------------
df13 = pd.read_sql("SELECT * FROM empleados WHERE salario > 3000;", engine)
print(df13)
```

```python
# -------------------------------
# 14. Listar las ventas realizadas en el canal 'WEB' durante 2024.
# -------------------------------
df14 = pd.read_sql("""
SELECT * FROM ventas
WHERE canal = 'WEB' AND DATEPART(YEAR, fecha_venta) = 2024;
""", engine)
print(df14)
```

```python
# -------------------------------
# 15. Mostrar las categorías de producto distintos.
# -------------------------------
df15 = pd.read_sql("SELECT categoria FROM productos GROUP BY categoria;", engine)
print(df15)
```

```python
# -------------------------------
# 16. Calcular el total de ventas por producto.
# -------------------------------
df16 = pd.read_sql("""
SELECT p.id_producto, p.nombre_producto,
       SUM(v.precio_unitario * v.cantidad) AS total_ventas
FROM productos p LEFT JOIN ventas v ON p.id_producto = v.id_producto
GROUP BY p.id_producto, p.nombre_producto
ORDER BY total_ventas DESC;
""", engine)
print(df16)
```

```python
# -------------------------------
# 17. Contar cuántos empleados tiene cada departamento.
# -------------------------------
df17 = pd.read_sql("""
SELECT d.id_departamento, COUNT(e.id_vendedor) AS Num_de_empleados
FROM departamentos d LEFT JOIN empleados e ON d.id_departamento = e.id_departamento
GROUP BY d.id_departamento
ORDER BY Num_de_empleados DESC;
""", engine)
print(df17)
```

```python
# -------------------------------
# 18. Mostrar las ventas agrupadas por región y canal.
# -------------------------------
df18 = pd.read_sql("""
SELECT region, canal, SUM(precio_unitario * cantidad) AS ventas_totales
FROM ventas
GROUP BY region, canal
ORDER BY region, ventas_totales DESC;
""", engine)
print(df18)
```

```python
# -------------------------------
# 19. Calcular el monto promedio de venta por cliente.
# -------------------------------
df19 = pd.read_sql("""
SELECT c.id_cliente, c.nombre,
       AVG(v.precio_unitario * v.cantidad) AS promedio_venta
FROM clientes c LEFT JOIN ventas v ON c.id_cliente = v.id_cliente
GROUP BY c.id_cliente, c.nombre
ORDER BY promedio_venta DESC;
""", engine)
print(df19)
```

```python
# -------------------------------
# 20. Obtener las fechas de la primera y última venta por cliente.
# -------------------------------
df20 = pd.read_sql("""
SELECT c.id_cliente,
       MAX(v.fecha_venta) AS fecha_última_compra,
       MIN(v.fecha_venta) AS fecha_primera_compra
FROM clientes c LEFT JOIN ventas v ON c.id_cliente = v.id_cliente
GROUP BY c.id_cliente
ORDER BY fecha_última_compra;
""", engine)
print(df20)
```

```python
# -------------------------------
# 21. Mostrar los clientes cuyo país es igual al país del cliente con id_cliente = 10.
# -------------------------------
df21 = pd.read_sql("""
SELECT c2.*
FROM clientes c2
WHERE c2.pais = (SELECT c.pais FROM clientes c WHERE c.id_cliente = 10);
""", engine)
print(df21)
```

```python
# -------------------------------
# 22. Listar los 5 productos más caros según precio_lista.
# -------------------------------
df22 = pd.read_sql("""
SELECT TOP (5) id_producto, nombre_producto, precio_lista
FROM productos
ORDER BY precio_lista DESC;
""", engine)
print(df22)
```

```python
# -------------------------------
# 23. Mostrar los clientes que no tienen edad registrada.
# -------------------------------
df23 = pd.read_sql("SELECT * FROM clientes WHERE edad IS NULL;", engine)
print(df23)
```

```python
# -------------------------------
# 24. Calcular el total de ventas del producto más vendido.
# -------------------------------
df24 = pd.read_sql("""
WITH t1 AS (
    SELECT TOP 1 v.id_producto, SUM(v.cantidad) AS cantidad_ventas
    FROM ventas v
    GROUP BY v.id_producto
    ORDER BY cantidad_ventas DESC
)
SELECT t1.id_producto,
       SUM(v1.precio_unitario * v1.cantidad) AS ventas
FROM t1 LEFT JOIN ventas v1 ON t1.id_producto = v1.id_producto
GROUP BY t1.id_producto;
""", engine)
print(df24)
```

```python
# -------------------------------
# 25. Mostrar las 5 ventas de mayor valor total.
# -------------------------------
df25 = pd.read_sql("""
SELECT TOP (5) id_venta,
       SUM(precio_unitario * cantidad) AS Total_venta
FROM ventas
GROUP BY id_venta
ORDER BY Total_venta DESC;
""", engine)
print(df25)
```
